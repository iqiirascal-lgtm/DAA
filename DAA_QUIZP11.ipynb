```python
# Soal 1 - Hitung Jumlah Angka Genap
def hitung_genap(arr):
    return sum(1 for x in arr if x % 2 == 0)

'''
Pendekatan/Algoritma:
- Fungsi melakukan penyisiran (traversal) terhadap seluruh elemen dalam array.
- Setiap elemen diuji menggunakan operasi modulus untuk menentukan apakah termasuk bilangan genap.
- Untuk setiap elemen yang memenuhi kriteria, nilai penghitung dinaikkan melalui mekanisme sum().

Edge cases:
- Jika array tidak memiliki elemen, proses penjumlahan tetap valid dan menghasilkan 0.
- Jika semua elemen merupakan bilangan ganjil, tidak ada nilai yang dihitung sebagai genap sehingga hasil tetap 0.
- Bilangan negatif juga diproses dengan aturan paritas yang sama, karena operasi modulus tetap berlaku.
'''

print(hitung_genap([1, 2, 3, 4, 7, 10]))
```

```python
# Soal 2 - Cari Nilai Terbesar
def cari_max(arr):
    if not arr:  # jika array kosong
        return None
    return max(arr)

'''
Langkah-langkah:
- Pemeriksaan awal dilakukan untuk memastikan bahwa array berisi data; jika tidak, fungsi mengembalikan None.
- Setelah lolos pengecekan, fungsi memanfaatkan max() untuk menentukan elemen dengan nilai paling tinggi.

Edge cases:
- Ketika array tidak memiliki elemen, fungsi wajib memberikan keluaran khusus agar tidak terjadi error.
- Jika seluruh nilai dalam array identik, nilai tersebut tetap dianggap sebagai nilai maksimum dan dikembalikan tanpa masalah.
'''
print(cari_max([5, 2, 9, 1]))
```

```python
# Soal 3 - Balik String
def balik_string(s):
    return s[::-1]

'''
Logika:
- Pemanfaatan slicing dengan langkah negatif memungkinkan string diakses dari indeks terakhir menuju awal.
- Teknik ini membentuk string baru yang susunan karakternya terbalik tanpa perlu perulangan eksplisit.
'''

print(balik_string("halo"))
```

```python
# Soal 4 - Jumlahkan Semua Elemen
def jumlahkan(arr):
    return sum(arr)

'''
Langkah-langkah:
- Seluruh elemen pada array dijumlahkan secara kumulatif oleh fungsi sum().
- Mekanisme ini secara internal melakukan penambahan berurutan hingga seluruh elemen terakumulasi.
'''

print(jumlahkan([3, 1, 2]))
```

```python
# Soal 5 - Cek Palindrome
def cek_palindrome(s):
    return s == s[::-1]

'''
Cara menentukan:
- String diperiksa dengan mencocokkannya terhadap versi yang telah dibalik.
- Jika kedua bentuk tersebut identik, maka string dikategorikan sebagai palindrome.
'''

print(cek_palindrome("level"))  # True
print(cek_palindrome("hello"))  # False
```

```python
# Soal 6 - Hitung Jumlah Huruf Vokal
def hitung_vokal(s):
    vokal = "aiueoAIUEO"
    return sum(1 for ch in s if ch in vokal)

'''
Cara memeriksa:
- Setiap karakter dievaluasi untuk mengetahui apakah termasuk dalam kumpulan huruf vokal.
- Jika karakter cocok, penghitung bertambah sehingga total vokal dapat diperoleh.
'''

print(hitung_vokal("informatika"))
```

```python
# Soal 7 - Temukan Elemen yang Dicari
def cari_target(arr, target):
    return target in arr

'''
Metode pencarian:
- Operator in melakukan pemeriksaan keberadaan elemen dengan memeriksa setiap posisi secara berurutan.
- Jika nilai ditemukan pada salah satu indeks, hasil pencarian bernilai True.
'''

print(cari_target([4, 7, 1, 9], 7))
```
